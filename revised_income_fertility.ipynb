```python
income_columns = [col for col in df.columns if 'income_' in col]
melted_df = df.melt(id_vars='State', value_vars=income_columns, var_name='Year', value_name='Income')

# Convert 'Year' to a numerical value (e.g., 'income_2010' to 2010)
melted_df['Year'] = melted_df['Year'].str.replace('income_', '').astype(int)

# Step 2: Plot the Data
plt.figure(figsize=(10, 6))
for state in melted_df['State'].unique():
    state_data = melted_df[melted_df['State'] == state]
    plt.plot(state_data['Year'], state_data['Income'], label=state)

plt.xlabel('Year')
plt.ylabel('Income')
plt.title('Income over Years by State')
plt.legend(title='State', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
# Step 1: Reshape the DataFrame for Fertility Rates
fertility_columns = [col for col in df.columns if 'FertilityRate_' in col]
melted_df = df.melt(id_vars='State', value_vars=fertility_columns, var_name='Year', value_name='FertilityRate')

# Convert 'Year' to a numerical value (e.g., 'FertilityRate_2010' to 2010)
melted_df['Year'] = melted_df['Year'].str.replace('FertilityRate_', '').astype(int)

# Step 2: Plot the Data for Fertility Rates
plt.figure(figsize=(10, 6))
for state in melted_df['State'].unique():
    state_data = melted_df[melted_df['State'] == state]
    plt.plot(state_data['Year'], state_data['FertilityRate'], label=state)

plt.xlabel('Year')
plt.ylabel('Fertility Rate')
plt.title('Fertility Rate over Years by State')
plt.legend(title='State', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()


```

```python
import matplotlib.pyplot as plt
import numpy as np

# Calculate change in fertility rate for each state
fertility_change = melted_df.groupby('State')['FertilityRate'].agg([np.min, np.max])
fertility_change['change'] = fertility_change['amax'] - fertility_change['amin']

# Identify states with the most and least change
most_change_states = fertility_change['change'].nlargest(2).index
least_change_states = fertility_change['change'].nsmallest(2).index

# Filter the melted_df for these states
filtered_states = most_change_states.union(least_change_states)
filtered_df = melted_df[melted_df['State'].isin(filtered_states)]

# Plot the data
plt.figure(figsize=(10, 6))
for state in filtered_df['State'].unique():
    state_data = filtered_df[filtered_df['State'] == state]
    plt.plot(state_data['Year'], state_data['FertilityRate'], label=state)

plt.xlabel('Year')
plt.ylabel('Fertility Rate')
plt.title('Fertility Rate Changes for States with Most and Least Change')
plt.legend(title='State')
plt.show()
```
